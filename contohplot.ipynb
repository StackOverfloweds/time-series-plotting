## Contoh Plotting Sederhana

### Apa yang Dilakukan?

    File ini berisi contoh dasar untuk membuat plot (grafik) menggunakan library matplotlib di Python.

    Kita membuat data sederhana secara manual, yaitu:

        Tanggal: Dari 1 November 2021 hingga 7 November 2021.

        Jumlah Kelas: Nilai acak seperti 5, 6, 8, 2, 3, 7, dan 4.

    Data ini disimpan dalam sebuah tabel (dataframe) menggunakan library pandas.

    Kemudian, kita membuat plot garis (line plot) yang menunjukkan hubungan antara tanggal dan jumlah kelas.

Tujuan:

    Memahami cara membuat plot dasar menggunakan matplotlib.

    Melihat bagaimana data time series (berdasarkan waktu) dapat divisualisasikan.

    Belajar mengatur judul, label sumbu, rotasi label, dan menambahkan grid pada plot.

Contoh Output:

    Sebuah grafik garis dengan:

        Sumbu X: Tanggal (1 Nov 2021 hingga 7 Nov 2021).

        Sumbu Y: Jumlah kelas.

        Garis biru yang menghubungkan titik-titik data.

        Judul: "Classes by Date".

        Grid untuk memudahkan pembacaan.

```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

### Create dataframe

```python
dataframe = pd.DataFrame({
    'date_of_week': np.array([datetime.datetime(2021, 11, i+1) for i in range(7)]),
    'classes': [5, 6, 8, 2, 3, 7, 4]
})
```

### Plotting the time series of given dataframe

```python
plt.figure(figsize=(10, 6))  # Mengatur ukuran plot
plt.plot(dataframe.date_of_week, dataframe.classes, marker='o', linestyle='-', color='b', label='Classes')
# Giving title to the chart using plt.title
plt.title('Classes by Date')
 
# rotating the x-axis tick labels at 30degree 
# towards right
plt.xticks(rotation=30, ha='right')
 
# Providing x and y label to the chart
plt.xlabel('Date')
plt.ylabel('Classes')
```
