# Plotting Dataset Time Series

### Apa yang Dilakukan?

File ini menggunakan dataset time series dari file CSV (DATA_Historical_Data.csv).

Dataset ini berisi data historis dengan kolom seperti:

        Date: Tanggal.

        Price: Harga pada tanggal tersebut.

        Open, High, Low: Harga pembukaan, tertinggi, dan terendah.

        Vol.: Volume perdagangan.

        Change %: Persentase perubahan harga.

Kita membaca dataset menggunakan pandas, mengubah kolom Date ke format tanggal yang benar, dan membuat plot time series.

Plot yang Dibuat:

    Plot Harga (Price) terhadap Waktu (Date):

        Menunjukkan bagaimana harga berubah dari waktu ke waktu.

        Garis hijau yang mewakili harga (Price).

    Plot Harga, Open, High, dan Low terhadap Waktu:

        Menampilkan beberapa variabel sekaligus (Price, Open, High, Low) dalam satu grafik.

        Setiap variabel memiliki warna berbeda untuk memudahkan pembacaan.

Tujuan:

    Memahami cara memvisualisasikan dataset time series.

    Melihat tren atau pola dari data historis (misalnya, bagaimana harga berubah dari waktu ke waktu).

    Belajar membandingkan beberapa variabel dalam satu plot.

Contoh Output:

    Grafik pertama:

        Sumbu X: Tanggal.

        Sumbu Y: Harga (Price).

        Garis hijau yang menunjukkan perubahan harga dari waktu ke waktu.

    Grafik kedua:

        Sumbu X: Tanggal.

        Sumbu Y: Nilai (Price, Open, High, Low).

        Empat garis dengan warna berbeda (hijau, biru, merah, oranye) yang mewakili masing-masing variabel.



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca dataset
df = pd.read_csv('DATA_Historical_Data.csv')
```

```python
# Mengubah kolom 'Date' ke format datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
```

```python
# Plotting harga (Price) terhadap waktu (Date)
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Price'], label='Price', color='green')
plt.title('Time Series Plot of Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting harga (Price), Open, High, dan Low terhadap waktu (Date)
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Price'], label='Price', color='green')
plt.plot(df['Date'], df['Open'], label='Open', color='blue')
plt.plot(df['Date'], df['High'], label='High', color='red')
plt.plot(df['Date'], df['Low'], label='Low', color='orange')
plt.title('Time Series Plot of Price, Open, High, and Low')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

## KESIMPULAN HASIL

1. Time Series Plot of Price (Plot Harga terhadap Waktu)

    Tren Harga: Plot menunjukkan perubahan harga (Price) dari waktu ke waktu. Terlihat bahwa harga mengalami fluktuasi, naik dan turun, dalam periode yang ditampilkan (dari 25 Mei 2023 hingga 5 Juni 2023).

    Pola: Ada beberapa titik di mana harga mencapai puncak (high) dan titik terendah (low), menunjukkan volatilitas dalam data.

    Kesimpulan: Harga tidak stabil dan cenderung berubah-ubah dalam periode ini. Hal ini bisa disebabkan oleh faktor eksternal seperti permintaan pasar, berita, atau faktor lainnya.

2. Time Series Plot of Price, Open, High, and Low (Plot Harga, Open, High, dan Low terhadap Waktu)

    Perbandingan Variabel:

        Harga (Price): Garis hijau menunjukkan harga penutupan pada setiap hari.

        Harga Pembukaan (Open): Garis biru menunjukkan harga pembukaan pada setiap hari.

        Harga Tertinggi (High): Garis merah menunjukkan harga tertinggi yang dicapai dalam sehari.

        Harga Terendah (Low): Garis oranye menunjukkan harga terendah dalam sehari.

    Pola:

        Terlihat bahwa High dan Low membentuk rentang harga harian, sementara Price dan Open berada di dalam rentang tersebut.

        Pada beberapa hari, Price (harga penutupan) mendekati High atau Low, menunjukkan pergerakan harga yang signifikan dalam sehari.

    Kesimpulan:

        Rentang antara High dan Low menunjukkan volatilitas harga harian.

        Price dan Open cenderung mengikuti pola yang sama, tetapi ada hari di mana perbedaan antara keduanya cukup signifikan, menunjukkan pergerakan harga yang kuat dalam sehari.
